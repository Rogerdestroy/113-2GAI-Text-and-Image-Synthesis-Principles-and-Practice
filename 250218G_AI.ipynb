#HW1在colab中畫一個函數圖形

```python
# 設定圖形大小
plt.figure(figsize=(8, 8))

# 設定點的數量(越多越圓滑)
n = 5000

# 設定 t 的範圍 (線段長度)
t = np.linspace(0, 6.26, n)

# 計算 x 和 y 座標
x = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12) ** 5)
y = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12) ** 5)

# 繪製蝴蝶曲線，使用紫色
plt.plot(x, y, color='#700080', lw=1)
# 設定標題
plt.title('Butterfly Curve')

# 設定 x 和 y 軸標籤
plt.xlabel('x')
plt.ylabel('y')

# 設定 x 和 y 軸的範圍
plt.xlim(-4, 4)
plt.ylim(-4, 4)

# 儲存圖形
plt.savefig('Butterfly_Curve.png')

# 顯示圖形
plt.show()
```

```python
def implicit_curve(x, y):
    return x**2 + (y - x**(2/3))**2 - 1

# 設定繪圖範圍
x_vals = np.linspace(-2, 2, 400)
y_vals = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = implicit_curve(X, Y)

# 繪製等高線圖來表示曲線
plt.contour(X, Y, Z, levels=[0], colors='b')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Implicit Curve: x^2 + (y - x^(2/3))^2 = 1")
plt.show()
```

```python
def koch_curve(n):
    if n == 0:
        return [[0, 0], [1, 0]]
    else:
        prev = koch_curve(n - 1)
        new = []
        for i in range(len(prev) - 1):
            new.append(prev[i])
            x1, y1 = prev[i]
            x2, y2 = prev[i + 1]
            x = (2 * x1 + x2) / 3
            y = (2 * y1 + y2) / 3
            new.append([x, y])
            x = (x1 + 2 * x2) / 3
            y = (y1 + 2 * y2) / 3
            new.append([x, y])
            x = (x1 + x2) / 2 + np.sqrt(3) * (y2 - y1) / 3
            y = (y1 + y2) / 2 - np.sqrt(3) * (x2 - x1) / 3
            new.append([x, y])
        new.append(prev[-1])
        return new

n = 4
curve = koch_curve(n)
curve = np.array(curve)
plt.figure(figsize=(10, 10))
plt.plot(curve[:, 0], curve[:, 1])
plt.axis('equal')
plt.axis('off')
plt.show()

```
